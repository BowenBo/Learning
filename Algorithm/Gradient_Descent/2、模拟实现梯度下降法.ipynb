# 模拟实现梯度下降法

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 141)
plot_y = (plot_x - 2.5) ** 2 -1
plt.plot(plot_x, plot_y)
```

```python
#求导数
def dJ(theta):
    return 2 * (theta - 2.5)
```

```python
# 损失函数
def J(theta):
    return (theta - 2.5) ** 2 -1
```

```python
#eta:学习率（theta移动步长）
eta = 0.1
#导数为0时是最低点，但精度不能达到导数刚好为0，因此设置一个判断到最低点的基本条件
epsilon = 1e-8
#theta：特征参数
theta = 0.0
while True:
    #导数gradient
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    if abs(J(theta) - J(last_theta)) < epsilon:
        break
print(theta)
print(J(theta))
print(dJ(theta))
```

```python
theta = 0.0
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    if abs(J(theta) - J(last_theta)) < epsilon:
        break
```

```python
plt.plot(plot_x, plot_y)
plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r', marker = '+')
```

### 算法封装

```python
def gradient_decent(initial_theta, eta, n_iters = 1e4, epsilon = 1e-8):
    '''
    initial_theta:初始theta
    eta：下降步长
    epsilon：极值，表示导数接近为0的时候到损失函数峰底
    n_iters：最多的循环次数，避免循环条件导致死循环
    '''
    
    theta = initial_theta
    theta_history.append(initial_theta)
    i_iters = 0
    while i_iters < n_iters:
        #下降梯度gradient（导数，斜率）
        gradient = dJ(theta)
        last_theta = theta#上一个theta
        theta = theta - eta * gradient#最新theta变化
        theta_history.append(theta)#theta列表
        
        if abs(J(theta) - J(last_theta)) < epsilon:
            break
        i_iters += 1

def plot_theta_history():
    plt.plot(plot_x, J(plot_x))
    plt.plot(np.array(theta_history), J(np.array(theta_history)), marker = '+')
```

```python
eta = 0.01
# eta越小，计算的点越多，eta太大可能会导致无导数
theta_history = []
gradient_decent(0.0, eta)
```

```python
plot_theta_history()
```
