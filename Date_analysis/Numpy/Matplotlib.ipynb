## 2、简单绘图
### 2.1、绘制sin()和cos（）函数
    * 获得sin cos函数数据 numpy
    * 绘图 pl.plot(x, y)
        - ipython --pylab   交互环境引入pylab
        - %pylab inline   在交货环境中直接绘图（否则可用pl.show()）

```python
import numpy as np
import pylab as pl
%pylab inline
x = np.linspace(-np.pi, np.pi, 256, endpoint = True)
C, S = np.sin(x), np.cos(x)
pl.plot(x, C)
pl.plot(x, S)
# pl.show()
```

### 2.2、默认值示例
    * 改变颜色和线宽度
        pl.plot(x, y, color="blue", linewidth= 2, linstyle="-")
     * 设置极值
         pl.xlim(x.min()*n, x.max()*m)
         pl.ylim(y.min()*n, y.max()*m)
     * 设置坐标轴刻度值
         现在的刻度不太理想，因为他们没有显示对于sine和cosine有意义的值（+/-π,+/-π/2）。我们将改变这些刻度，让他们只显示这些值.
         pl.xticks(-np.pi, -np.pi/2, 0, np.pi/2, np.pi)
     * 设置刻度标签
     

```python
#     * 创建一个大小为 8X6 英寸figsize，每英寸80个点的图片dpi
pl.figure(figsize = (8, 6), dpi = 80)

#     * 从1X1的网格创建一个子图片
pl.subplot(1, 1, 1)

#     * 用宽度为1（像素）的蓝色连续直线绘制cosine
pl.plot(x, C, color = 'blue', linewidth = 1.0, linestyle = '-')

#     * 用宽度为1（像素）的绿色连续直线绘制sine
pl.plot(x, S, color = 'green', linewidth = '1.0', linestyle ='-')

#     * 设置x轴的极值
# pl.xlim(-4.0, 4.0)

#     * 当前的图片的极值限制太拥挤了，我们希望留一点空间以便清晰的看到所有的数据点
pl.xlim(x.min() * 1.1,  x.max() * 1.1)
pl.ylim(C.min() * 1.1, C.max() * 1.1)

#     * 设置x轴的刻度值
pl.xticks(np.linspace(-4, 4, 9, endpoint = True))

#     * 设置y轴的极值
# pl.ylim(-1.0, 1.0)

#     * 设置y轴的刻度值
pl.yticks(np.linspace(-1, 1, 5,  endpoint = True))

#     * 用72dpi保存图片
savefig('exercise_1.png', dpi = 72)
```
