```python
import numpy as np
import matplotlib.pyplot as plt
```

## 移动坐标系脊柱
    * ax = plt.gca
    * ax.spines['right/left/bottom/top'].set_color()   #'none'可以隐藏脊柱
    * ax.yaxis.set_ticks_position('left')#将左脊柱设置为y轴线
    * ax.xaxis.set_ticks_position('bottom')#将下脊柱设置为x轴线
    * ax.spines[''].set_position('data', 0) #将某脊柱移动到data为某值的位置
## 设置极值
    *plt.xlim(start, end)
    *plt.ylim(start, end)    
## 设置线条颜色、宽度、形状
    * plt.plot(x,y,color='',linewidth=n,linestyle='')
## 添加图例
    * plt.plot(x,y,label='')
    * plt.legend(loc='upper/lower left/right')
## 设置坐标轴刻度值、标题
    * plt.xticks([])
    * plt.yticks([])
    * plt.title('')

```python
x = np.arange(-10, 10, 0.1)
u = np.e ** x
u1 = np.e ** (-x)
sh_x = (np.e ** x + np.e ** (-x))/2
ch_x = (np.e ** x - np.e ** (-x))/2
plt.figure(figsize = (8, 6), dpi = 160)
plt.subplot(1,1,1)
plt.xlim(-4,4)
plt.ylim(-5,5)
#设置坐标
ax = plt.gca()  #get current axis
ax.spines['right'].set_color("none")  #spine脊骨
ax.spines['top'].set_color("none")
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
plt.xticks([-4, -2, 0, 2, 4])
plt.yticks([-5, -4, -3, -2, -1, 1, 2, 3, 4, 5])
plt.xlabel('x')
plt.title('双曲函数')
plt.plot(x, u/2, color='red', linewidth=1, linestyle='-.', label='$e^x/2$')
plt.plot(x, u1/2, color='blue', linewidth=1, linestyle='-.', label='$e^{-x}/2$')
plt.plot(x, u + u1, color='green', linewidth=2, linestyle='-', label=r"$ \frac{(e^{x}+e^{-x})}{2}$")
plt.plot(x, u - u1, color='pink', linewidth=2, linestyle='-', label=r"$ \frac{(e^{x}-e^{-x})}{2}$")
plt.legend(loc='upper left')
```

## toolkits.axisartist绘制坐标轴及箭头

```python
import mpl_toolkits.axisartist as axisartist
#画布设置
fig = plt.figure(dpi=120)
#位置定义
ax1 = axisartist.Subplot(fig, 111)
#位置中添加轴
fig.add_axes(ax1)
#绘制箭头
ax1.axis['bottom'].set_axisline_style('-|>', size=1.5)
ax1.axis['left'].set_axisline_style('-|>', size=1.5)
#隐藏上、右边
ax1.axis['top'].set_visible(False)
ax1.axis['right'].set_visible(False)
```
